```python
# 2) Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release)and make data frame.
```

```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
page=requests.get('https://www.imdb.com/chart/top')
```

```python
soup=BeautifulSoup(page.content)
```

```python
names=[]
for i in soup.find_all('td',class_="titleColumn"):
    i=i.text.replace(("\n")," ")
    i=i.strip(" ")
    names.append(i)     
```

```python
ratings=[]
for i in soup.find_all('td',class_="ratingColumn imdbRating"):
    i=i.text.replace("\n",' ')
    i=i.strip('')
    ratings.append(i)
```

```python
yearofrelease=[]
for i in soup.find_all('span',class_="secondaryInfo"):
    i=i.text.replace("\n",' ')
    i=i.strip('')
    yearofrelease.append(i)
```

```python
import pandas as pd
df=pd.DataFrame({'MovieName':names,'yearof release':yearofrelease,'ratings':ratings},index=range(0,250))
df[0:100]
```

```python
# 5) Write a python program to scrape cricket rankings from icc-cricket.com. You have to scrape:
#a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
import pandas as pd
table=soup.find('table',attrs={"class":"table"})
header=[]

for i in table.find_all('th'):
    title=i.text.strip()
    header.append(title)
    df=pd.DataFrame(columns=header)
    
for row in table.find_all('tr')[1:]:
    data=row.find_all('td')
    row_data=[td.text.replace('\n',"").strip() for td in data]
    length=len(df)
    df.loc[length]=row_data
df[0:10]
```

```python
# 7) Write a python program to scrape details of all the posts from coreyms.com. Scrape the heading, date, content and the code for the video from the link for the youtube video from the post.
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://coreyms.com/development/')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
headings=[]
for i in soup.find_all('h2',class_="entry-title"):
    headings.append(i.text)
```

```python
date=[]
for i in soup.find_all('time',class_="entry-time"):
    date.append(i.text)
```

```python
content=[]
for i in soup.find_all('div',class_="entry-content"):
    i=i.text.replace("\n",' ')
    i=i.strip('')
    content.append(i)
```

```python
video=[]
for i in soup.find_all('iframe',class_='youtube-player'):
    video.append(i['src'])
```

```python
import pandas as pd
df=pd.DataFrame({'Heading':headings,'Date':date,'Content':content,'Video':video})
df
```

```python
# 9) Write a python program to scrape mentioned details from dineout.co.in :
#i) Restaurant name
#ii) Cuisine
#iii) Location
#iv) Ratings
#v) Image URL 
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
titles=[]
for i in soup.find_all('div',class_='restnt-info cursor'):
    titles.append(i.text)
```

```python
location=[]
for i in soup.find_all('div',class_="restnt-loc ellipsis"):
    location.append(i.text)
```

```python
cuisine=[]
for i in soup.find_all('span',class_="double-line-ellipsis"):
    cuisine.append(i.text)
```

```python
ratings=[]
for i in soup.find_all('div',class_="restnt-rating rating-4"):
    ratings.append(i.text)
```

```python
images=[]
for i in soup.find_all('img',class_="no-img"):
    images.append(i['data-src'])
```

```python
import pandas as pd
df=pd.DataFrame({'Restaurent Name':titles,'Cuisines':cuisine,'Location':location,'Ratings':ratings,'Images_url':images})
df
```
