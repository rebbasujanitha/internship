B) Name
C) Artist
D) Upload date
E) Views

```python
driver=webdriver.Chrome(r'C:\Users\polasasuresh\Downloads\chromedriver_win32 (1)\chromedriver.exe')
```

```python
driver.get('https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos')
```

```python
#creating an empty list
Rank=[]
Name=[]
Artist=[]
uploaddate=[]
views=[]
```

```python
#scraping Rank details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[1]")[:30]:
        Rank.append(i.text)
except NoSuchElementException:
    Rank.append("-")
#scraping Name details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[2]")[:30]:
        Name.append(i.text)
except NoSuchElementException:
    Name.append("-")
#scraping Artist details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[3]")[:30]:
        Artist.append(i.text)
except NoSuchElementException:
    Artist.append("-")
#scraping uploaddate details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[5]")[:30]:
        uploaddate.append(i.text)
except NoSuchElementException:
    uploaddate.append("-")
#scraping views details
try:
    for i in driver.find_elements_by_xpath("//table[@class='wikitable sortable jquery-tablesorter']//tbody//tr//td[4]")[:30]:
        views.append(i.text)
except NoSuchElementException:
    views.append("-") 
```

```python
df=pd.DataFrame({"Rank":Rank,"Name":Name,"Artist":Artist,"uploaddate":uploaddate,"views":views})
df
```

```python
#clcikcing viewalldatasets tab
driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b").click()
```

```python
#scraping Name of the dataset
Name=[]
for i in driver.find_elements_by_xpath("//table[@border='1']/tbody/tr/td[1]//p//a"):
    Name.append(i.text)

#scraping Task of the dataset
Task=[]
for i in driver.find_elements_by_xpath("//table[@border='1']//tbody/tr/td[3]/p"):
    try:
        Task.append(i.text)
    except NoSuchElementException:
        Task.append("-")

#scraping AttributeType
AttributeType=[]
for i in driver.find_elements_by_xpath("//table[@border='1']//tbody/tr/td[4]/p"):
    try:
        AttributeType.append(i.text)
    except NoSuchElementException:
        AttributeType.append("-")

#scraping NoofInstances
Instances=[]
for i in driver.find_elements_by_xpath("//table[@border='1']//tbody/tr/td[5]/p"):
    try:
        Instances.append(i.text)
    except NoSuchElementException:
        Instances.append("-")

#scraping NoofAttributes
NoAttributes=[]
for i in driver.find_elements_by_xpath("//table[@border='1']//tbody/tr/td[6]/p"):
    try:
        NoAttributes.append(i.text)
    except NoSuchElementException:
        NoAttributes.append("-")

#scraping year
Year=[]
for i in driver.find_elements_by_xpath("//table[@border='1']//tbody/tr/td[7]/p"):
    try:
        Year.append(i.text)
    except NoSuchElementException:
        Year.append("-")

```

```python
df=pd.DataFrame({"DatasetName":Name,"Task":Task,"No.Of.Instances":Instances,"AttributeType":AttributeType,"No.Of.Attributes":NoAttributes,"Year":Year})
df
```
